```
using System;
using System.Collections.Generic;

interface IWarranty
{
    void Warranty();
}

public abstract class Supplier
{
    public string Name;
    public string Site;
    public string Number;
    public string Email;
    public string Address;
    public string ContactPerson;
    public string ProductType;  // Новый атрибут
    public string TIN;         // Новый атрибут (код налогоплательщика)
    public static int TotalSupplier { get; private set; } = 0;

    public Supplier(string name, string site, string number, string email, string address, string contactPerson, string productType, string tin)
    {
        Name = name;
        Site = site;
        Number = number;
        Email = email;
        Address = address;
        ContactPerson = contactPerson;
        ProductType = productType;
        TIN = tin;
        TotalSupplier++;
    }

    public static void NumberOfSupplier()
    {
        Console.WriteLine($"Общее количество поставщиков: {TotalSupplier}");
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Название поставщика - {Name}.");
        Console.WriteLine($"Сайт компании - {Site}.");
        Console.WriteLine($"Контактный номер телефона - {Number}.");
        Console.WriteLine($"Адрес - {Address}.");
        Console.WriteLine($"Email - {Email}.");
        Console.WriteLine($"Контактное лицо - {ContactPerson}.");
        Console.WriteLine($"Тип продукции - {ProductType}.");
        Console.WriteLine($"Код налогоплательщика (TIN) - {TIN}.");
    }

    public double CalculateShippingCost(string region)
    {
        // Примерная логика расчета стоимости доставки
        double cost = region.ToLower() switch
        {
            "москва" => 500,
            "питер" => 400,
            _ => 700
        };
        return cost;
    }
}

public class InternetionalSuplier : Supplier
{
    public string Country;
    public string DeliveryRegion;  // Новый атрибут для региона доставки

    public InternetionalSuplier(string name, string site, string number, string email, string address, string contactPerson, string productType, string tin, string country, string deliveryRegion)
        : base(name, site, number, email, address, contactPerson, productType, tin)
    {
        Country = country;
        DeliveryRegion = deliveryRegion;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Страна поставки - {Country}.");
        Console.WriteLine($"Регион доставки - {DeliveryRegion}.");
    }

    public double CalculateDeliveryTime()
    {
        // Примерная логика расчета времени доставки
        return DeliveryRegion.ToLower() switch
        {
            "москва" => 2,
            "питер" => 3,
            _ => 5
        };
    }
}

public class Food : InternetionalSuplier
{
    public int ExpirationDate;
    public string StorageConditions;
    public string Certification; // Новый атрибут для сертификатов

    public Food(string name, string site, string number, string email, string address, string contactPerson, string productType, string tin, string country, string deliveryRegion, int expirationDate, string storageConditions, string certification)
        : base(name, site, number, email, address, contactPerson, productType, tin, country, deliveryRegion)
    {
        ExpirationDate = expirationDate;
        StorageConditions = storageConditions;
        Certification = certification;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Срок годности продуктов на складе - {ExpirationDate} лет.");
        Console.WriteLine($"Условия хранения - {StorageConditions}.");
        Console.WriteLine($"Сертификаты качества - {Certification}.");
    }

    public bool CheckCertifications()
    {
        return !string.IsNullOrEmpty(Certification);
    }
}

public class BuildingMaterial : InternetionalSuplier
{
    public double Weight;
    public string Type;
    public double PricePerUnit;
    public string MaterialCategory; // Новый атрибут для категории материала

    public BuildingMaterial(string name, string site, string number, string email, string address, string contactPerson, string productType, string tin, string country, string deliveryRegion, double weight, string type, double pricePerUnit, string materialCategory)
        : base(name, site, number, email, address, contactPerson, productType, tin, country, deliveryRegion)
    {
        Weight = weight;
        Type = type;
        PricePerUnit = pricePerUnit;
        MaterialCategory = materialCategory;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Вес строительного материала - {Weight} кг.");
        Console.WriteLine($"Тип строительного материала - {Type}.");
        Console.WriteLine($"Цена за единицу - {PricePerUnit} руб.");
        Console.WriteLine($"Категория материала - {MaterialCategory}.");
    }

    public bool IsEcoFriendly()
    {
        return MaterialCategory.ToLower() == "экологичный";
    }
}

public class Electronics : InternetionalSuplier, IWarranty
{
    public string TypeElectronics;
    public int WarrantyElectronics;
    public double Price;
    public string Manufacturer; // Новый атрибут для производителя

    public Electronics(string name, string site, string number, string email, string address, string contactPerson, string productType, string tin, string country, string deliveryRegion, string type, int warranty, double price, string manufacturer)
        : base(name, site, number, email, address, contactPerson, productType, tin, country, deliveryRegion)
    {
        TypeElectronics = type;
        WarrantyElectronics = warranty;
        Price = price;
        Manufacturer = manufacturer;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Тип устройств на складе - {TypeElectronics}.");
        Console.WriteLine($"Цена устройства - {Price} руб.");
        Console.WriteLine($"Производитель устройства - {Manufacturer}.");
    }

    // Явная реализация метода интерфейса IWarranty
    void IWarranty.Warranty()
    {
        Console.WriteLine($"Гарантия устройств составляет - {WarrantyElectronics} год.");
    }

    public bool IsCompatibleWith(string deviceType)
    {
        return TypeElectronics.Contains(deviceType);
    }
}

public class SupplierCollection<T> where T : Supplier
{
    private List<T> _suppliers = new List<T>();

    public void Add(T supplier)
    {
        _suppliers.Add(supplier);
    }

    public void Remove(T supplier)
    {
        _suppliers.Remove(supplier);
    }

    public void DisplayAll()
    {
        foreach (var supplier in _suppliers)
        {
            supplier.GetInfo();
            Console.WriteLine();
        }
    }
}

// Пример использования
SupplierCollection<Supplier> supplierCollection = new SupplierCollection<Supplier>();

InternetionalSuplier inter = new InternetionalSuplier("TranStar", "www.transtar.ru", "8938678453", "transtar.123@mail.ru", "Комсомольская 60", "Тимур", "Товары", "1234567890", "Россия", "Москва");
supplierCollection.Add(inter);

Food foo = new Food("Кабан72", "www.kabanfood72.ru", "89265788334", "kaban.72@mail.ru", "Широтная 45", "Влад", "Продукты", "0987654321", "Россия", "Питер", 1, "Температура: 0-4°C", "ISO 9001");
supplierCollection.Add(foo);

BuildingMaterial bildmat = new BuildingMaterial("Строй Дом", "www.stroydom.ru", "89653458723", "stroy.huse@mail.ru", "Кремлевская 85", "Алексей", "Строительные материалы", "1122334455", "Россия", "Москва", 1254.5, "Дерево", 1500, "Экологичный");
supplierCollection.Add(bildmat);

Electronics elec = new Electronics("Ток", "www.tok.ru", "89386784532", "electron.123@mail.ru", "Пермякова 60", "Андрей", "Электроника", "6677889900", "Россия", "Питер", "Телевизор", 1, 30000, "Samsung");
supplierCollection.Add(elec);

supplierCollection.DisplayAll();
Supplier.NumberOfSupplier();

Console.ReadLine();
```
