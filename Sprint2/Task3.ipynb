Создать базовый класс Supplier в C#, который будет представлять информацию о
поставщиках товаров или услуг. На основе этого класса разработать 2-3
производных класса, демонстрирующих принципы наследования и полиморфизма.
В каждом из классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
using System;

    interface IWarranty
    {
        void Warranty();
    }
    public abstract class Supplier
    {
        public string Name;
        public string Site;
        public string Number;
        public string Email;
        public string Address;
        public static int TotalSupplier { get; private set; } = 0;

        public Supplier(string name, string site, string number, string email,string address)
        {
            Name = name;
            Site = site;
            Number = number;
            Email = email;
            Address = address;
            TotalSupplier++;
        }
        public static void NumberOfSupplier()
        {
            Console.WriteLine($"Общее количество поставщиков: {TotalSupplier}");
        }

    }    
    public class InternetionalSuplier : Supplier
    {
        public string Country;
        public InternetionalSuplier(string name, string site, string number, string email, string address, string country): base(name,site,number,email,address)
        {
            Country = country;
        }

        public virtual void GetInfo()
        {
            Console.WriteLine($"Название поставщика - {Name}.");
            Console.WriteLine($"Сайт компании - {Site}.");
            Console.WriteLine($"Контактный номер телефона - {Number}.");
            Console.WriteLine($"Адрес - {Address}.");
            Console.WriteLine($"Email - {Email}.");
            Console.WriteLine($"Страна поставки - {Country}.");
        }
        public void ReplaceName(string name)
        {
            Name = name;
        }
        public void PrintNumber()
        {
            Console.WriteLine($"Номер телефона - {Number}.");
        }
    }

    public class Food : InternetionalSuplier
    {
        public int ExpirationDate;

        public Food(string name, string site, string number, string email, string address, string country, int date) : base(name, site, number, email, address, country)
        {
            ExpirationDate = date;
        }
        public override void GetInfo()
        {
            base.GetInfo();
            Console.WriteLine($"Срок годности продуктов на складе - {ExpirationDate} лет.");
        }
        public void OverDate()
        {
            int result = DateTime.Now.Year;
            result += ExpirationDate;

            Console.WriteLine($"Срок годности закончится в {result} году.");
        }
    }

    public class BuildingMaterial : InternetionalSuplier
    {
        public double Weight;
        public string Type;

        public BuildingMaterial(string name, string site, string number, string email, string address, string country, double weight, string type) : base(name, site, number, email, address, country)
        {
            Weight = weight;
            Type = type;
        }
        public override void GetInfo()
        {
            base.GetInfo();
            Console.WriteLine($"Вес строительного материала - {Weight} кг.");
            Console.WriteLine($"Тип строительного материала - {Type}.");
        }
        public void NecessaryWeight()
        {
            if (Weight < 1300.0)
            {
                Console.WriteLine($"Данный материал подходит для транспортировки.");
            }
            else
            {
                Console.WriteLine($"Данный материал не подходит для транспортировки.");
            }
        }
    }
    public class Electronics : InternetionalSuplier, IWarranty
    {
        public string TypeElectronics;
        public int WarrantyElectronics;
        public Electronics(string name, string site, string number, string email, string address, string country, string type, int warranty) : base(name, site, number, email, address, country)
        {
            TypeElectronics = type;
            WarrantyElectronics = warranty;
        }
        public override void GetInfo()
        {
            base.GetInfo();
            Console.WriteLine($"Тип устройств на складе - {TypeElectronics}.");
        }
        public void Warranty()
        {
            Console.WriteLine($"Гарантия устройств составляет - {WarrantyElectronics} год.");
        }

    }
    InternetionalSuplier inter = new InternetionalSuplier("TranStar", "www.transtar.ru", "8938678453", "Комсомольская 60", "transtar.123@mail.ru", "Россия");
    inter.GetInfo();
    Console.WriteLine();

    Food foo = new Food("Кабан72", "www.kabanfood72.ru", "89265788334", "Широтная 45", "kaban.72@mail.ru", "Россия", 1);
    foo.GetInfo();
    foo.OverDate();
    Console.WriteLine();

    BuildingMaterial bildmat = new BuildingMaterial("Строй Дом", "www.stroydom.ru", "89653458723", "Кремлевская 85", "stroy.huse@mail.ru", "Россия", 1254.5, "Дерево");
    bildmat.GetInfo();
    bildmat.NecessaryWeight();
    Console.WriteLine();

    Electronics elec = new Electronics("Ток", "www.tok.ru", "89386784532", "Пермякова 60", "electron.123@mail.ru", "Россия", "Телевизор", 1);
    elec.GetInfo();
    elec.Warranty();
    Console.WriteLine();
    Supplier.NumberOfSupplier();

    Console.ReadLine();
```
