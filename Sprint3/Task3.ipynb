```

using System;
using System.Collections.Generic;

interface IWarranty
{
    void Warranty();
}

public abstract class Supplier
{
    public string Name;
    public string Site;
    public string Number;
    public string Email;
    public string Address;
    public string ContactPerson;
    public static int TotalSupplier { get; private set; } = 0;

    public Supplier(string name, string site, string number, string email, string address, string contactPerson)
    {
        Name = name;
        Site = site;
        Number = number;
        Email = email;
        Address = address;
        ContactPerson = contactPerson;
        TotalSupplier++;
    }

    public static void NumberOfSupplier()
    {
        Console.WriteLine($"Общее количество поставщиков: {TotalSupplier}");
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Название поставщика - {Name}.");
        Console.WriteLine($"Сайт компании - {Site}.");
        Console.WriteLine($"Контактный номер телефона - {Number}.");
        Console.WriteLine($"Адрес - {Address}.");
        Console.WriteLine($"Email - {Email}.");
        Console.WriteLine($"Контактное лицо - {ContactPerson}.");
    }
}

public class InternetionalSuplier : Supplier
{
    public string Country;

    public InternetionalSuplier(string name, string site, string number, string email, string address, string contactPerson, string country)
        : base(name, site, number, email, address, contactPerson)
    {
        Country = country;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Страна поставки - {Country}.");
    }

    public void GetInfo(bool includeCountry)
    {
        base.GetInfo();
        if (includeCountry)
        {
            Console.WriteLine($"Страна поставки - {Country}.");
        }
    }
}

public class Food : InternetionalSuplier
{
    public int ExpirationDate;
    public string StorageConditions;

    public Food(string name, string site, string number, string email, string address, string contactPerson, string country, int expirationDate, string storageConditions)
        : base(name, site, number, email, address, contactPerson, country)
    {
        ExpirationDate = expirationDate;
        StorageConditions = storageConditions;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Срок годности продуктов на складе - {ExpirationDate} лет.");
        Console.WriteLine($"Условия хранения - {StorageConditions}.");
    }

    public void GetInfo(int maxExpiration)
    {
        base.GetInfo();
        if (ExpirationDate <= maxExpiration)
        {
            Console.WriteLine($"Срок годности продуктов на складе - {ExpirationDate} лет.");
        }
        else
        {
            Console.WriteLine($"Продукты с более чем {maxExpiration} летним сроком годности не отображаются.");
        }
    }
}

public class BuildingMaterial : InternetionalSuplier
{
    public double Weight;
    public string Type;
    public double PricePerUnit;

    public BuildingMaterial(string name, string site, string number, string email, string address, string contactPerson, string country, double weight, string type, double pricePerUnit)
        : base(name, site, number, email, address, contactPerson, country)
    {
        Weight = weight;
        Type = type;
        PricePerUnit = pricePerUnit;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Вес строительного материала - {Weight} кг.");
        Console.WriteLine($"Тип строительного материала - {Type}.");
        Console.WriteLine($"Цена за единицу - {PricePerUnit} руб.");
    }

    public void GetInfo(double maxPrice)
    {
        base.GetInfo();
        if (PricePerUnit <= maxPrice)
        {
            Console.WriteLine($"Вес строительного материала - {Weight} кг.");
            Console.WriteLine($"Тип строительного материала - {Type}.");
            Console.WriteLine($"Цена за единицу - {PricePerUnit} руб.");
        }
        else
        {
            Console.WriteLine("Этот строительный материал слишком дорогой.");
        }
    }
}

public class Electronics : InternetionalSuplier, IWarranty
{
    public string TypeElectronics;
    public int WarrantyElectronics;
    public double Price;

    public Electronics(string name, string site, string number, string email, string address, string contactPerson, string country, string type, int warranty, double price)
        : base(name, site, number, email, address, contactPerson, country)
    {
        TypeElectronics = type;
        WarrantyElectronics = warranty;
        Price = price;
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Тип устройств на складе - {TypeElectronics}.");
        Console.WriteLine($"Цена устройства - {Price} руб.");
    }

    public void Warranty()
    {
        Console.WriteLine($"Гарантия устройств составляет - {WarrantyElectronics} год.");
    }

    public void GetInfo(double maxPrice)
    {
        base.GetInfo();
        if (Price <= maxPrice)
        {
            Console.WriteLine($"Тип устройств на складе - {TypeElectronics}.");
            Console.WriteLine($"Цена устройства - {Price} руб.");
        }
        else
        {
            Console.WriteLine("Это устройство слишком дорогое.");
        }
    }
}

public class SupplierCollection<T> where T : Supplier
{
    private List<T> _suppliers = new List<T>();

    public void Add(T supplier)
    {
        _suppliers.Add(supplier);
    }

    public void Remove(T supplier)
    {
        _suppliers.Remove(supplier);
    }

    public void DisplayAll()
    {
        foreach (var supplier in _suppliers)
        {
            supplier.GetInfo();
            Console.WriteLine();
        }
    }
}
SupplierCollection<Supplier> supplierCollection = new SupplierCollection<Supplier>();

InternetionalSuplier inter = new InternetionalSuplier("TranStar", "www.transtar.ru", "8938678453", "transtar.123@mail.ru", "Комсомольская 60", "Тимур", "Россия");
supplierCollection.Add(inter);

Food foo = new Food("Кабан72", "www.kabanfood72.ru", "89265788334", "kaban.72@mail.ru", "Широтная 45", "Влад", "Россия", 1, "Температура: 0-4°C");
supplierCollection.Add(foo);

BuildingMaterial bildmat = new BuildingMaterial("Строй Дом", "www.stroydom.ru", "89653458723", "stroy.huse@mail.ru", "Кремлевская 85", "Алексей", "Россия", 1254.5, "Дерево", 1500);
supplierCollection.Add(bildmat);

Electronics elec = new Electronics("Ток", "www.tok.ru", "89386784532", "electron.123@mail.ru", "Пермякова 60","Андрей", "Россия", "Телевизор", 1, 30000);
supplierCollection.Add(elec);

supplierCollection.DisplayAll();
Supplier.NumberOfSupplier();

Console.ReadLine();
```
